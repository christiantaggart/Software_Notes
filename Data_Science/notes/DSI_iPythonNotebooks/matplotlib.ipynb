```python
import numpy as np
from numpy import random
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
xs = random.normal(0, .001, 50)
ys = random.normal(0, 1, 50)
fig = plt.figure(figsize=(5, 5))
ax = fig.add_subplot(1, 1, 1)
ax.scatter(x=xs, y=ys)
ax.set_xlabel('xs', size=15)
ax.set_ylabel('ys', size=15)
```

```python
xs = np.linspace(-2., 2., 20)
ys = xs ** 2
fig = plt.figure(figsize=(5, 5))
ax = fig.add_subplot(1, 1, 1)
ax.plot(xs, ys)
ax.set_xlabel('x', size=15)
ax.set_ylabel('x^2', size=15)
```

```python
xs = np.linspace(-2., 2., 15)
ys = xs ** 3
fig = plt.figure(figsize=(10, 5))
ax = fig.add_subplot(1, 1, 1)
ax.plot(xs, ys, 'r:') # b- == blue line ro == red dot
ax.set_xlabel('x', size=15)
ax.set_ylabel('x^2', size=15)
ax.set_title("TITLE HERE")
```

```python
xs = np.linspace(-2., 2., 100)
ys = xs ** 3
fig = plt.figure(figsize=(10, 5))
ax = fig.add_subplot(1, 1, 1)
ax.plot(xs, ys, 'r-.', label='cubic')
ax.set_xlabel('x', size=15)
ax.set_ylabel('x^3', size=15)
ax.legend()
```

```python
xs = np.linspace(-10., 10., 100)
ys = xs ** 3
fig = plt.figure(figsize=(10, 5))
ax = fig.add_subplot(1, 1, 1)
ax.plot(xs, ys, label='cubic', lw=1, alpha=0.5, marker='^', markersize=7)# marker size = dot size
ax.set_xlabel('x', size=15)
ax.set_ylabel('x^3', size=15)
ax.legend()
```

```python
orders = range(1, 4)  # Orders of polynomails
colors = ['r', 'g', 'b']
labels = ['linear', 'quadratic', 'cubic']
markers = ['+', 'o', '*']
lss = ['--', '-', ':']
lws = [6, 3, 2]
alpha = 0.5

xs = np.linspace(-2., 2., 10)

fig = plt.figure(figsize=(10, 5))
ax = fig.add_subplot(1, 1, 1)
ax.set_xlabel('x')
ax.set_ylabel('y')

for order, color, label, marker, ls, lw in zip(orders, colors, labels, markers, lss, lws):
    ys = xs ** order
    ax.plot(xs, ys, label=label, color=color, marker=marker, markersize=lw * 2, ls=ls, lw=lw, alpha=alpha)

ax.legend(loc='best')
```

```python
orders = range(1, 4)  # Orders of polynomails
colors = ['r', 'g', 'b']
titles = ['linear', 'quadratic', 'cubic']

xs = np.linspace(-2., 2., 100)

# Tuple unpacking
fig, axes = plt.subplots(nrows=1, ncols=len(orders), figsize=(10,3))

plt.tight_layout( pad=5)


for ax, order, color, title in zip(axes, orders, colors, titles):
    ys = xs ** order
    ax.set_title(title)
    ax.set_xlabel('x')
    ax.set_ylabel('x^{}'.format(order))
    ax.plot(xs, ys, color)
```

```python
number_of_walks = 5
steps_in_each_walk = 100

# Make random distances for each step
step_distances = random.randn(number_of_walks * (steps_in_each_walk -1)).reshape(5, -1)

# Make each walk start at step 0 by horizontally stacking 0s
# at the beginning of the walks

starts = np.zeros(step_distances.shape[0]).reshape(-1, 1)
walk_distances = np.hstack([starts, step_distances])
walks = walk_distances.cumsum(axis=1)

# Now make an array of integers that will serve as the x axis
steps = np.arange(walks.shape[1])

# Find the mean distance at the end of each walk by taking the mean
# of the last column of walks

mean_dist = walks[:, -1].mean()

# Colors, labels, and alpha settings for our lines
colors = ['r', 'g', 'b', 'm', 'c']
labels = ['A', 'B', 'C', 'D', 'E']
alpha = 0.5

# Create a grid of 1 by 1 plots and size it
fig = plt.figure(figsize=(15, 5))
ax = fig.add_subplot(1, 1, 1)

# Make the axis labels
ax.set_ylabel('distance')
ax.set_xlabel('steps')

# Force the x axis limits to match the first and last step
# by default matplotlib may try to pad this which will make
# it look like a mess

ax.set_xlim(steps[0], steps[-1])

# Plot each walk. Zip together the iterables we will need and
# just plot to the same axes

for walk_idx, color, label in zip(range(walks.shape[0]), colors, labels):
    ax.plot(steps, walks[walk_idx], label=label, color=color, lw=1)

# Behind every great plot is an inspirational legend.
# Make your here

ax.legend(loc='best')

# Make a line (and label for it) that show the mean distance
# convered by each walk

ax.axhline(mean_dist, 0, steps[-1])
ax.text(steps[-1] + 1, mean_dist, 'μ={:.2f}'.format(mean_dist))
```
